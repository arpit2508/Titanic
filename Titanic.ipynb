```python
from sklearn import tree
```

```python
train= pd.read_csv("D:/Downloads/titanic/train.csv")
```

```python
train.head()
```

```python
train["Fare"]=train["Fare"].fillna(train["Fare"].dropna().median())
train["Age"]=train["Age"].fillna(train["Age"].dropna().median())
train.loc[train["Sex"]=="male", "Sex"] = 0
train.loc[train["Sex"]=="female", "Sex"]= 1

train["Embarked"]= train["Embarked"].fillna("S")
train.loc[train["Embarked"]=="S","Embarked"]=0
train.loc[train["Embarked"]=="C","Embarked"]=1
train.loc[train["Embarked"]=="Q","Embarked"]=2
```

```python
train.head()
```

```python
target= train["Survived"].values
feature_names=["Pclass","Age","Fare","Embarked", "Sex","SibSp","Parch"]
features= train[feature_names].values

decision_tree= tree.DecisionTreeClassifier(random_state = 1)
decision_tree_results= decision_tree.fit(features, target)
print(decision_tree_results.score(features, target))
```

```python
from sklearn import model_selection
```

```python
scores = model_selection.cross_val_score(decision_tree, features, target, scoring='accuracy', cv=50)
print(scores)
print(scores.mean())
```

```python
dot_data=tree.export_graphviz(decision_tree_results, feature_names=feature_names, out_file="tree.dot"  )
```
